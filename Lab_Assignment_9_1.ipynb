TASK-1: Auto-Generating Google Style Function Documentation

Prompt: Give me Python code for creating multiple functions and also provide Google-style documentation in docstrings including description, parameters, return values, and example usage.

```python
def add(a, b):
    """
    Adds two numbers and returns the result.

    Args:
        a (int | float): First number.
        b (int | float): Second number.

    Returns:
        int | float: Sum of the two numbers.

    Example:
        >>> add(3, 5)
        8
    """
    return a + b


def factorial(n):
    """
    Calculates factorial of a non-negative integer.

    Args:
        n (int): Non-negative integer.

    Returns:
        int: Factorial value of n.
    """
    if n == 0:
        return 1
    return n * factorial(n - 1)


def is_prime(num):
    """
    Checks whether a number is prime.

    Args:
        num (int): Number to check.

    Returns:
        bool: True if prime, False otherwise.
    """
    if num < 2:
        return False
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False
    return True


# Driver Code
print(add(3, 5))
print(factorial(5))
print(is_prime(11))
```

Observation: The Google-style docstrings clearly explain the purpose, parameters, return type, and usage of each function.
New developers can understand the functionality without reading the implementation logic.
The code becomes self-documenting and professional.

TASK-2
Adding Inline Comments to Complex Logic

Prompt: Add inline comments to the Python code explaining only complex or non-obvious logic such as algorithms, loops, and conditionals.

```python
def fibonacci(n):
    fib_seq = [0, 1]

    # Generate sequence iteratively to avoid recursion overhead
    for i in range(2, n):
        fib_seq.append(fib_seq[i - 1] + fib_seq[i - 2])

    return fib_seq[:n]


def binary_search(arr, target):
    left, right = 0, len(arr) - 1

    while left <= right:
        # Calculate midpoint safely
        mid = (left + right) // 2

        if arr[mid] == target:
            return mid

        # If target is greater, ignore left half
        elif arr[mid] < target:
            left = mid + 1

        # Otherwise ignore right half
        else:
            right = mid - 1

    return -1


print(fibonacci(7))
print(binary_search([10, 20, 30, 40, 50], 40))
```

Observation: Inline comments were added only around algorithmic logic and tricky conditions.
The difficult parts of the code became easier to follow without cluttering the function body.
This improves readability for beginners.

TASK-3: Generating Module-Level Documentation
Prompt: Generate a professional module-level docstring for the Python file that describes the module’s purpose, dependencies, key functions, and example usage without modifying the code.

```python
"""
math_utils.py

This module provides basic mathematical utilities
such as factorial calculation, prime checking,
Fibonacci sequence generation, and mean calculation.

Dependencies:
    math (standard library)

Example:
    >>> import math_utils
    >>> math_utils.factorial(5)
    120
"""

import math


def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n - 1)


def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(math.sqrt(num)) + 1):
        if num % i == 0:
            return False
    return True


def mean(numbers):
    return sum(numbers) / len(numbers)


print(factorial(5))
print(is_prime(7))
print(mean([10, 20, 30]))
```

Observation: The module-level docstring provides an immediate overview of the module’s purpose and usage.
Anyone opening the file can quickly understand what the module does.
This improves usability and maintainability.

TASK-4: Converting Inline Comments to Structured Docstrings
Prompt: Convert detailed inline comments inside each function into structured Google-style docstrings while preserving functionality.

```python
def calculate_discount(price, discount_percent):
    """
    Calculates final price after applying percentage discount.

    Args:
        price (float): Original price.
        discount_percent (float): Discount percentage.

    Returns:
        float: Final discounted price.
    """
    discount_amount = (discount_percent / 100) * price
    return price - discount_amount


def find_max(numbers):
    """
    Finds maximum value in a list.

    Args:
        numbers (list[int | float]): List of numeric values.

    Returns:
        int | float: Maximum value.
    """
    if not numbers:
        raise ValueError("List cannot be empty")

    max_value = numbers[0]
    for num in numbers:
        if num > max_value:
            max_value = num

    return max_value


print(calculate_discount(100, 20))
print(find_max([10, 45, 23, 67, 12]))
```

Observation: The inline comments were transformed into clean and standardized docstrings.
Redundant comments were removed, making the code more readable and professional.
Consistency across the codebase improved significantly.

TASK-5: Mini Automatic Documentation Generator

Prompt: Create a Python utility that reads a .py file, detects all functions and classes, and automatically inserts placeholder Google-style docstrings where missing, without altering existing documentation or functionality.

```python
import ast
import os

FUNCTION_DOC = """Brief description of the function.

Args:
    TODO: Describe parameters.

Returns:
    TODO: Describe return value.
"""

class DocstringInserter(ast.NodeTransformer):

    def visit_FunctionDef(self, node):
        if ast.get_docstring(node) is None:
            node.body.insert(0, ast.Expr(value=ast.Constant(value=FUNCTION_DOC)))
        return self.generic_visit(node)


def generate_docstrings(input_file, output_file):

    if not os.path.exists(input_file):
        print("File not found")
        return

    with open(input_file, "r") as f:
        source = f.read()

    tree = ast.parse(source)
    updated_tree = DocstringInserter().visit(tree)
    ast.fix_missing_locations(updated_tree)

    updated_code = ast.unparse(updated_tree)

    with open(output_file, "w") as f:
        f.write(updated_code)

    print("Docstrings added successfully!")


# Example usage
generate_docstrings("task1.py", "task1_with_docs.py")
```

Observation: The tool successfully detects functions using Python’s AST module.
Placeholder Google-style docstrings are inserted only where missing.
The original code behavior remains unchanged.
This utility improves documentation adoption and maintainability.
