```python
# Task 1 - Syntax Error Fix

def greet():
    return "Hello, AI Debugging Lab!"

print(greet())

# Test cases
assert greet() == "Hello, AI Debugging Lab!"
assert len(greet()) > 0
assert "Hello" in greet()
```

```python
# Task 2 - Incorrect If Condition Fix

def check_number(n):
    if n == 10:
        return "Ten"
    else:
        return "Not Ten"

print(check_number(10))
print(check_number(5))

# Test cases
assert check_number(10) == "Ten"
assert check_number(5) == "Not Ten"
assert check_number(0) == "Not Ten"
```

```python
# Task 3 - File Not Found Fix

def read_file(filename):
    try:
        with open(filename, "r") as f:
            return f.read()
    except FileNotFoundError:
        return f"Error: The file '{filename}' was not found. Please check the file path and try again."

print(read_file("nonexistent.txt"))

# Test cases
assert "Error" in read_file("nonexistent.txt")
assert "not found" in read_file("nonexistent.txt")
```

```python
# Task 4 - Fix missing method in class

class Car:
    def start(self):
        return "Car started"

    def drive(self):
        return "Car is driving"

my_car = Car()
print(my_car.drive())

# Test case
assert my_car.start() == "Car started"
```

```python
# Task 5 - TypeError Fix

# Solution 1: Type casting (convert string to int)
def add_five(value):
    if isinstance(value, str):
        value = int(value)
    return value + 5

# Solution 2: String concatenation
def add_five_concat(value):
    if isinstance(value, str):
        return value + "5"
    else:
        return str(value) + "5"

print(add_five("10"))          # 15
print(add_five_concat("10"))   # "105"

# Test cases for add_five
assert add_five(10) == 15
assert add_five("10") == 15
assert add_five(5) == 10

# Test cases for add_five_concat
assert add_five_concat("10") == "105"
assert add_five_concat(10) == "105"
assert add_five_concat("hello") == "hello5"
```
